```python
#Standardize header names (lower cases + _)
```

```python
df1.columns = df1.columns.str.lower().str.replace(' ', '_')
df2.columns = df2.columns.str.lower().str.replace(' ', '_')
df3.columns = df3.columns.str.lower().str.replace(' ', '_')
```

```python
#Standardize header names (same names)
```

```python
df1 = df1.rename(columns={'st': 'state'})
df2 = df2.rename(columns={'st': 'state'})
```

```python
#Rearrange the columns in the dataframe as needed
```

```python
df1.columns
```

```python
df1 = df1[['customer','gender' ,'state', 'education',
       'income', 'customer_lifetime_value', 'monthly_premium_auto', 'number_of_open_complaints',
       'policy_type', 'vehicle_class', 'total_claim_amount']]
```

```python
df2 = df2[['customer','gender' ,'state', 'education',
       'income', 'customer_lifetime_value', 'monthly_premium_auto', 'number_of_open_complaints',
       'policy_type', 'vehicle_class', 'total_claim_amount']]
```

```python
df3 = df3[['customer','gender' ,'state', 'education',
       'income', 'customer_lifetime_value', 'monthly_premium_auto', 'number_of_open_complaints',
       'policy_type', 'vehicle_class', 'total_claim_amount']]
```

```python
#Correct the values in the column customer lifetime value. They are supposed to be absolute numbers...

df1["customer_lifetime_value"] = df1["customer_lifetime_value"].str.replace('%','')
df2["customer_lifetime_value"] = df2["customer_lifetime_value"].str.replace('%','')
```

```python
#Concatenate the three dataframes
```

```python
lst_to_concat = [df1,df2,df3] 
new_table = pd.concat(lst_to_concat)
new_table
```
